```python
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Matches.csv')
```

```python
#Having a glance at first five records of dataset.
df.head()
```

```python
#Looking at the number of columns and rows of dataset.
df.shape
```

```python
#Lokking for non null count and dtypes 
df.info()
```

```python
df.dropna(inplace = True)
```

```python
#aChecking for null values in dataset
df.isnull().sum()
```

```python
#Getting the frequency of most man of the match awards.
df['player_of_match'].value_counts()
```

```python
#Getting the top 10 player with most man of the match award
df['player_of_match'].value_counts()[:10]
```

```python
plt.figure(figsize=(15,5))
plt.bar(list(df['player_of_match'].value_counts()[:10].keys()),list(df['player_of_match'].value_counts()[:10]),color = 'r')
plt.title('Top 10 Player',fontsize=20)
plt.xlabel('Cricketer')
plt.ylabel('Number of Man of the Match won')
plt.show()
```

```python
#Getting the frequency of Result column 
df['result'].value_counts()
```

```python
#Finding the number of toss wins with respect to each team 
df['toss_winner'].value_counts() 
```

```python
#Extracting all the records where team won batting first 
batting_first = df[df['result']!='wickets']
batting_first
```

```python
plt.figure(figsize=(10,5))
plt.hist(batting_first['result_margin'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.ylabel('Number Of Matches')
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first 
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(6,5))
plt.bar(list(batting_first['winner'].value_counts().keys()[0:3]),list(batting_first['winner'].value_counts()[0:3]),color = 'g')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Top 3 Team (after batting first)')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(batting_first['winner'].value_counts()), labels = list(batting_first['winner'].value_counts().keys()), autopct = '%0.1f%%')
plt.show()
```

```python
#Extracting all the records where team won batting second 
batting_second = df[df['result']!='runs']
batting_second
```

```python
plt.figure(figsize=(10,5))
plt.hist(batting_second['result_margin'])
plt.title('Distribution of Wickets')
plt.xlabel('Wickets')
plt.ylabel('Number Of Matches')
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting second 
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(12,5))
plt.bar(list(batting_second['winner'].value_counts().keys()[0:5]),list(batting_second['winner'].value_counts()[0:5]),color = 'g')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.title('Top 3 Team (after batting second)')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(batting_second['winner'].value_counts()), labels = list(batting_second['winner'].value_counts().keys()), autopct = '%0.1f%%')
plt.show()
```

```python
df['Seasons'] = pd.DatetimeIndex(df.date).year
```

```python
#Looking for the number of matches played each season
df['Seasons'].value_counts()
```

```python
#Looking For the number of matches played in each city 
df['city'].value_counts()
```

```python
#Looking for the team who won the match after winning the toss 
x = df[df['toss_winner']==df['winner']]
x
```

```python
x['winner'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.bar(list(x['winner'].value_counts().keys()),list(x['winner'].value_counts()))
plt.title('Distribution of Team(Team who won the match after winning toss)')
plt.ylabel('Matches Won')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Looking for the team who won the match after lossing the toss 
y = df[df['toss_winner']!=df['winner']]
y
```

```python
y['winner'].value_counts()
```

```python
plt.figure(figsize=(10,8))
plt.bar(list(y['winner'].value_counts().keys()),list(y['winner'].value_counts()))
plt.title('Distribution of Team(Team who won the match after lossing toss)')
plt.ylabel('Matches Won')
plt.xticks(rotation = 90)
plt.show()
```
